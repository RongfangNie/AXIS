### 选择方差前10000个基因


```python
import pandas as pd
import numpy as np



omics_dict = {"mRNA": "OmicsExpressionProteinCodingGenesTPMLogp1BatchCorrected.csv",
            "Gene dependency": "CRISPRGeneDependency.csv",
            "Gene effect": "CRISPRGeneEffect.csv",
            "ssGSEA": "OmicsExpressionGeneSetEnrichment.csv",
            "DNA Methylation": "Methylation_(1kb_upstream_TSS)_subsetted_NAsdropped.csv",
            "CNV": "OmicsCNGene.csv",
            }


omic = "mRNA"
data_dir = f"./data/Depmap/{omics_dict[omic]}"

```

```python
GDSC_data = pd.read_csv(data_dir,index_col=0).T
GDSC_data = GDSC_data.dropna()
gene_var = GDSC_data.var(axis=1)
gene_var.sort_values(ascending=False)
gene_var.index = GDSC_data.index
var_gene = gene_var.sort_values(ascending=False)[:10000,].index
```

## 选择高变基因的数据

```python
var_gene_data = GDSC_data[GDSC_data.index.isin(var_gene)]
```

## 标准化


```python
from scipy.stats import zscore
z_data = pd.DataFrame(zscore(var_gene_data,axis=1), columns=var_gene_data.columns, index=var_gene_data.index)
```

```python
z_data.var(axis=1).sort_values(ascending=False)
```

## 划分测试集和训练集

```python

from sklearn.model_selection import train_test_split
z_dataT = z_data.T
z_data_train, z_data_test = train_test_split(z_dataT, test_size=0.2, random_state=42)
```

```python
out_dir_dict = {"mRNA": "Depmap_ProteinCodingGene_mRNA",
            "Gene dependency": "Depmap_gene_dependency",
            "Gene effect": "Depmap_gene_effect",
            "ssGSEA": "Depmap_ssGSEA",
            "DNA Methylation": "Depmap_methlation",
            "CNV": "Depmap_CNV",
            }
```
