### 选择方差前10000个基因


```python
import pandas as pd
import numpy as np
data_dir = "./data/Depmap/OmicsSomaticMutationsMatrixDamaging.csv"
```

```python
GDSC_data = pd.read_csv(data_dir,index_col=0)
GDSC_data = GDSC_data.T
GDSC_data = GDSC_data.dropna()
gene_var = GDSC_data.var(axis=1)
gene_var.index = GDSC_data.index
var_gene = gene_var.sort_values(ascending=False)[:256,].index
```

## 选择高变基因的数据

```python
var_gene_data = GDSC_data[GDSC_data.index.isin(var_gene)]
```

## 标准化


```python
from scipy.stats import zscore
#z_data = pd.DataFrame(zscore(var_gene_data,axis=1), columns=var_gene_data.columns, index=var_gene_data.index)
z_data = var_gene_data
z_data = z_data.T
```
