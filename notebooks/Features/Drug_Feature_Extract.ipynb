````python
import pandas as pd

sm_ls = pd.read_pickle("../../data/DrugComb/Process/DrugComb_all_witch_CID_data_CID_2_SMILES_dict.pkl")
````

````python
import torch
print(torch.Tensor)  # 应该输出 <class 'torch.Tensor'>
import torch
print(torch.__version__)
````

````python
# 提取 keys 和 values 为两个列表
keys_list = list(sm_ls.keys())
values_list = list(sm_ls.values())
````

````python
import datamol as dm
from molfeat.trans.pretrained.hf_transformers import PretrainedHFTransformer

# sanitize and standardize your molecules if needed

transformer = PretrainedHFTransformer(kind='Roberta-Zinc480M-102M', notation='smiles', dtype=float)
````

````python
features = transformer(values_list)
````

````python
cid_features = {}
for i in range(len(keys_list)):
    cid = keys_list[i]
    feature = features[i]
    cid_features[cid] = features[i]
````

````python
import pickle

# 指定保存的文件路径
file_path = '../../data/DrugComb/Process/cid_features_Roberta.pkl'

# 使用pickle模块将字典保存为pkl文件
with open(file_path, 'wb') as file:
    pickle.dump(cid_features, file)
````
